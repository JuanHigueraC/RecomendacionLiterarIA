# Limpeza de datos

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/books_google_dataset.csv')
```

```python
def create_quality_table(df):
    quality_table = pd.DataFrame({
        'Column': df.columns,
        'Data Type': df.dtypes,
        'Count': df.count(),
        'Missing Values': df.isnull().sum(),
        'Missing %': (df.isnull().sum() / len(df)) * 100,
        'Unique Values': df.nunique(),
        'Unique %': (df.nunique() / len(df)) * 100
    })
    
    # Round percentages to 2 decimal places
    quality_table['Missing %'] = quality_table['Missing %'].round(2)
    quality_table['Unique %'] = quality_table['Unique %'].round(2)
    
    return quality_table
```

```python
# Create and display the quality table
quality_table = create_quality_table(df)
display(quality_table)
```

```python
# For multiple columns using a list
columns_to_check = ['isbn13', 'title', 'authors', 'categories', 'description', 'published_year', 'num_pages']
df.dropna(subset=columns_to_check, inplace=True)
```

```python
# Create and display the quality table
quality_table = create_quality_table(df)
display(quality_table)
```
